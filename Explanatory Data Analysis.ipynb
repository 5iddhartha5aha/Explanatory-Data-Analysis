```python
import numpy as np
import scipy
import matplotlib.pyplot
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
# Reading CSV file
data = pd.read_csv("weatherAUS 2.csv")
```

```python
data
```

```python
# Converting values of columns : RainToday & RainTomorrow
data.replace('Yes', 1, inplace=True)
data.replace('No', 0, inplace=True)
```

```python
data
```

```python
# Converting date values to numeric ones
data['Date'] = pd.to_datetime(data['Date'])

# Choose a reference date (epoch)
reference_date = pd.Timestamp('2008-12-01')

# Convert the date column to the number of days since the reference date
data['date_as_float'] = (data['Date'] - reference_date).dt.total_seconds() / (24 * 60 * 60)
```

```python
data
```

```python
# Dropping the original date column
data.drop('Date', axis=1, inplace=True)
```

```python
data
```

```python
# applying onehot encoding to convert categorical values to numeric ones
onehot_encoded_data = pd.get_dummies(data, columns = ['Location', 'WindGustDir','WindDir9am','WindDir3pm'])
onehot_encoded_data
```

```python
# replacing the null values with the median of their respective features
for i in range(len(onehot_encoded_data.columns)):
    onehot_encoded_data[onehot_encoded_data.columns[i]].fillna(onehot_encoded_data[onehot_encoded_data.columns[i]].median(), inplace=True)
```

```python
# to check for any null values left
onehot_encoded_data.isnull().sum()
```

```python
# scaling the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(onehot_encoded_data)
```

```python
scaled_data
```

```python
# applying dimensionality reduction to choose the number of dimensions that add up to a sufficiently large proportion of the variance, let's say 95%.
pca = PCA(n_components = 0.95)
pca.fit(scaled_data)
reduced_data = pca.fit_transform(scaled_data)
```
